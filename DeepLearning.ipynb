```python
#
import numpy as np
def AND(x1,x2):
    w = np.array([0.5,0.5])
    x = np.array([x1,x2])
    b = -0.7
    a = x*w
    tmp = np.sum(a) + b
    if tmp < 0:
        return 0
    else:
        return 1
def NAND(x1,x2):
    w = np.array([-0.5,-0.5])
    x = np.array([x1,x2])
    b = 0.7
    a = x*w
    tmp = np.sum(a) + b
    if tmp < 0:
        return 0
    else:
        return 1
def OR(x1,x2):
    w = np.array([0.5,0.5])
    x = np.array([x1,x2])
    b = -0.2
    a = x*w
    tmp = np.sum(a) + b
    if tmp < 0:
        return 0
    else:
        return 1
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    return AND(s1,s2)
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
#NumPy  x  0  True 0  False y
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
#
def step_function(x):
    y = x > 0
    # y  int  0  1  y  int 
    #astype()  NumPy astype()  np.int 
    return y.astype(np.int)
x = np.arange(-3,3,0.1)
y = step_function(x)
plt.plot(x,y,label = 'step_function')
plt.ylim(-0.1,1.1)
plt.legend()
plt.show()
#sigmoid 
def sigmoid(x):
    return 1/(1 + np.exp(-x))
x = np.arange(-10,10,0.1)
y = sigmoid(x)
plt.plot(x,y,label = 'sigmoid')
plt.legend()
plt.show()
#ReLU
def ReLU(x):
    return np.maximum(0,x)
x = np.arange(-10,10,0.1)
y = ReLU(x)
plt.plot(x,y,label = 'ReLU')
plt.legend()
plt.show()
def softmax(x):
    c = np.max(x)
    exp_a = np.exp(x-c)
    y = exp_a/np.sum(exp_a)
    return y
x = np.arange(-10,10,0.1)
y = softmax(x)
plt.plot(x,y,label = 'softmax')
plt.legend()
plt.show()
x = np.arange(-6,6,0.1)
y1 = step_function(x)
y2 = sigmoid(x)
y3 = ReLU(x)
y4 = softmax(x)
plt.plot(x,y1,label = 'step_function')
plt.plot(x,y2,label = 'simoid',linestyle = '-.')
plt.plot(x,y3,label = 'ReLU',linestyle = '--')
plt.plot(x,y4,label = 'softmax',linestyle = '--')
plt.legend()
plt.show()
```

```python
import os,sys
import numpy as np
from common.functions import *
from common.gradient import numerical_gradient
class TwoLayerNet:
    def __init__(self,input_size,hide_size,output_size,weight_init_std=0.01):
        self.params = {}
        self.params['w1'] = weight_init_std*(np.random.randn(input_size,hide_size))
        self.params['b1'] = np.zeros(hide_size)
        self.params['w2'] = weight_init_std*(np.random.randn(input_size,hide_size))
        self.params['b2'] = np.zeros(output_size)
    def predict(self,x):
        w1 = params['w1']
        w2 = params['w2']
        b1 = params['b1']
        b2 = params['b2']
        
        z1 = np.dot(x,w1) + b1
        z2 = sigmoid(z1)
        z3 = np.dot(z2,w2) + b2
        y = softmax(z3)
        
        return y
    
    def loss(self, x, t):
        y = self.predict(x)
        return cross_entropy_error(y, t)
    
    def accuracy(self, x, t):
        y = self.predict(x)
        y = np.argmax(y, axis=1)
        t = np.argmax(t, axis=1)
        return np.sum(y == t)/float(x.shape[0])
    
    def numerical_gradient(self, x, t):
        loss_W = lambda W: self.loss(x, t)

        grads = {}
        grads['w1'] = numerical_gradient(loss_W, self.params['w1'])
        grads['b1'] = numerical_gradient(loss_W, self.params['b1'])
        grads['w2'] = numerical_gradient(loss_W, self.params['w2'])
        grads['b2'] = numerical_gradient(loss_W, self.params['b2'])

        return grads
net = TwoLayerNet(input_size=784,hide_size=50,output_size=10)
print(net.params['w1'])
print(net.params['w2'])
print(net.params['b1'])
print(net.params['b2'])
```
